```python
import pandas as pd
import numpy as np
```

```python
# create sample dataset
new_sample=False
if new_sample:
    df = pd.read_csv('data-in/traffic-3xways', names=['Time', 'VID', 'Spd', 'XWay', 'Lane', 'Dir', 'Seg', 'Pos'])
    vids = df.VID.unique()
    df_sample = df[df.VID.isin(np.random.choice(vids, 100000))]
    df_sample = df_sample.sort_values(by=['Time'])
    df_sample.to_csv('data-in/traffic_sample_100kvids_new', index=False, header=False)
else:
    df_sample=pd.read_csv('data-in/traffic_sample_100kvids', names=['Time', 'VID', 'Spd', 'XWay', 'Lane', 'Dir', 'Seg', 'Pos'])
```

```python
# examples of uncompleted segments
dfg = df_sample[(df_sample.Seg>=52) & (df_sample.Seg <= 56)].groupby(by='VID').agg({'Seg': ['min', 'max']})
# examples of typical uncompleted segments
mask=((dfg['Seg']['min']==52) & (dfg['Seg']['max']!=56))
dfg[mask].head(5)
```

```python
# more specific examples of uncompleted segments
mask=((dfg['Seg']['min']==56) | (dfg['Seg']['max']==52))
dfg[mask].head(5)
```

```python
# examples of cars with 2 trips of different direction within 52 and 56 segments
dfg = df_sample[(df_sample.Seg>=52) & (df_sample.Seg <= 56)].groupby(by='VID').agg({'Dir': ['min', 'max']})
mask=(dfg['Dir']['min']!=dfg['Dir']['max'])
dfg[mask].head(10)
```
